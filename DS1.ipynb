```python
import csv

# Read the CSV file into a list of rows
with open('total_and_rev.csv', 'r') as f:
    reader = csv.reader(f)
    rows = list(reader)

# Extract the head of the CSV file (the first few rows)
head = rows[:5]

# Extract the tail of the CSV file (the last few rows)
tail = rows[-5:]


print('Head:')
for row in head:
    print(row)
print('\nTail:')
for row in tail:
    print(row)
```
