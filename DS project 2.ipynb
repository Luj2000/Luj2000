```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file and store the data in a Pandas dataframe
df = pd.read_csv("/content/stores_data_full_clean.csv")

# Set the style of the plot to 'bmh'
plt.style.use('bmh')

# Select the columns to plot
x = df['Date']
y = df['TotalPrice']

# Create the plot
plt.plot(x, y)

# Add a title and axis labels
plt.title('Total Sales')
plt.xlabel('Date')
plt.ylabel('Total Price')

# Show the plot
plt.show()
```

```python
import numpy as np

# Generate some data to plot
lon_grid, lat_grid = np.meshgrid(np.linspace(-180, 180, 20), np.linspace(-90, 90, 20))
temp_anomaly = np.sin(lon_grid) * np.cos(lat_grid)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate some data to plot
X, Y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
Z = np.sin(X) * np.cos(Y)

# Create the figure and subplot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Add a surface plot to the subplot
ax.plot_surface(X, Y, Z, cmap='Reds')

# Show the plot
plt.show()
```

```python
import seaborn as sns
```

```python
x = df['Date']
y = df['Revenue']

# Set the seaborn style
sns.set()

# Create the figure and subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Add a line plot to the subplot
sns.lineplot(x, y, ax=ax)

# Add a title to the subplot
ax.set_title('Richland Income Based on Age')

# Add an x-axis label to the subplot
ax.set_xlabel('Age', labelpad=15)
```

```python
sns.set()

# Create the figure and subplot
fig, ax = plt.subplots(figsize=(10, 5))

# Add a bar plot to the subplot
sns.barplot(x='Branch', y='Revenue', data=df, ax=ax)

# Add a title to the subplot
ax.set_title('Revenue by Branch')

# Add an x-axis label to the subplot
ax.set_xlabel('Branch', labelpad=15)

# Add a y-axis label to the subplot
ax.set_ylabel('Revenue (USD)', labelpad=15)

# Show the plot
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

slice_2019 = df[df['Date'].dt.year == 2019]

# Set the seaborn style
sns.set()

# Create the figure and subplot
fig, ax = plt.subplots(figsize=(10, 10))

# Add a heatmap to the subplot
sns.heatmap(slice_2019.corr(), annot=True, ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Add a regression plot to the subplot
sns.regplot(x='Quantity', y='TotalPrice', data=df, ax=ax)

plt.show()
```
